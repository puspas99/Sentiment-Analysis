```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv("clean_data_dec05.csv")

sns.countplot(x='sentiment', data=df)
```

```python
class0 = " ".join(text for text in df[df["sentiment"]=='Storngly Positive'].Tweet)
class1 = " ".join(text for text in df[df["sentiment"]=='Storngly Nagative'].Tweet)
class2 = " ".join(text for text in df[df["sentiment"]=='Positive'].Tweet)
class3 = " ".join(text for text in df[df["sentiment"]=='Nagative'].Tweet)
class3 = " ".join(text for text in df[df["sentiment"]=='Neutral'].Tweet)
```

```python
wordcloud = WordCloud(stopwords = STOPWORDS, background_color = "white", max_words = 1000).generate(class0)
plt.figure(figsize = (12, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.title("Storngly Nagative Wordcloud") 
```
